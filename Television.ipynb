### Number of Products with Full HD resolution

```python
num_of_full_hd = data['Picture_qualtiy'].str.contains("Full HD").sum()
total_products = data.shape[0]
print("There are total {} products with Full HD resolution.".format(num_of_full_hd))
print("Which is {:.2f} % of total products.".format((num_of_full_hd / total_products) * 100))
```

### Number of Products with Ultra HD(4K) resolution

```python
num_of_ultra_hd = data['Picture_qualtiy'].str.contains("Ultra HD").sum()

print("There are total {} products with Ultra HD(4k) resolution.".format(num_of_ultra_hd))
print("Which is {:.2f} % of total products.".format((num_of_ultra_hd / total_products) * 100))
```
